```python
def find_min_max_and_indices(lst):
    if not lst:
        return None, None, None, None
    min_index = max_index = 0
    min_val = max_val = lst[0]
    for i, num in enumerate(lst):
        if num < min_val:
            min_val = num
            min_index = i
        elif num > max_val:
            max_val = num
            max_index = i
    return min_val, max_val, min_index, max_index

CAN_min_val, CAN_max_val, CAN_min_index, CAN_max_index = find_min_max_and_indices(CAN_analysis)
FRA_min_val, FRA_max_val, FRA_min_index, FRA_max_index = find_min_max_and_indices(FRA_analysis)
DEU_min_val, DEU_max_val, DEU_min_index, DEU_max_index = find_min_max_and_indices(DEU_analysis)
ITA_min_val, ITA_max_val, ITA_min_index, ITA_max_index = find_min_max_and_indices(ITA_analysis)
JPN_min_val, JPN_max_val, JPN_min_index, JPN_max_index = find_min_max_and_indices(JPN_analysis)
KOR_min_val, KOR_max_val, KOR_min_index, KOR_max_index = find_min_max_and_indices(KOR_analysis)
UK_min_val, UK_max_val, UK_min_index, UK_max_index = find_min_max_and_indices(UK_analysis)
USA_min_val, USA_max_val, USA_min_index, USA_max_index = find_min_max_and_indices(USA_analysis)
```


불경기를 증명할 때 사용되는 데이터는 여러 가지가 있으며, 주로 경기 순환의 여러 측면을 포함하는 다양한 지표를 고려해야 합니다. 주요한 데이터와 지표에는 다음과 같은 것들이 있습니다:

실업률: 불경기 기간 동안 일반적으로 실업률이 상승합니다. 실업률이 지속적으로 높은 수준을 유지하거나 급격히 상승하는 것은 경기 침체를 나타낼 수 있습니다.

경기 침체 지표: 경기 침체를 측정하기 위한 여러 지표가 있습니다. 예를 들어, GDP 성장률이 음수인 경우, 생산량이 줄어드는 것을 나타내므로 경기 침체의 징후로 볼 수 있습니다.

소매판매 및 소비 지표: 소매판매량의 감소나 소비지출의 축소는 경기 침체의 징후로 여겨질 수 있습니다.

주택 시장 지표: 주택 판매량의 감소, 주택 가격의 하락 및 주택건설 활동의 감소는 경기 침체를 나타낼 수 있습니다.

기업 수익과 생산성: 기업의 이익이 감소하고 생산성이 저하되는 것은 경기 침체의 지표일 수 있습니다.

금융 시장 지표: 주가지수의 하락, 신용 스프레드의 확대, 이자율의 하락 등이 경기 침체의 징후로 여겨질 수 있습니다.

물가 지표: 소비자 물가 지수(CPI)나 생산자 물가 지수(PPI)의 하락은 수요 부진을 나타낼 수 있습니다.

이러한 데이터와 지표들을 종합적으로 분석하여 경기의 상황을 평가하고, 그 결과를 통해 불경기 여부를 판단할 수 있습니다. 하지만, 이러한 지표들을 해석할 때는 여러 요인을 고려해야 하며, 단일 지표만으로는 완전한 판단을 내리기에는 부족할 수 있습니다.

가지고 있는 CPI (소비자 물가 지수), 이자율 및 실업률 데이터를 활용하여 다양한 경제 지표 및 추이에 대한 분석을 수행하여 보고서를 작성할 수 있습니다. 아래는 가능한 보고서의 몇 가지 예시입니다:

경제 상황 분석 보고서:

이 보고서에서는 CPI, 이자율 및 실업률의 최신 데이터를 분석하여 현재의 경제 상황을 평가할 수 있습니다.
CPI의 증가는 소비자 물가 상승을 나타내며, 이자율의 변화는 경기에 미치는 영향을 반영할 수 있습니다.
실업률의 변화는 노동 시장의 건강 상태를 나타내며, 경제의 성장 가능성에 영향을 미칩니다.
인플레이션 및 금융정책 분석 보고서:

CPI와 이자율 데이터를 사용하여 인플레이션 추세 및 이에 대한 금융 정책의 영향을 분석할 수 있습니다.
CPI의 상승은 통화량 확대, 수요 증가 또는 공급 제약 등과 관련된 다양한 요인에 따라 발생할 수 있습니다.
이러한 인플레이션 추세에 대응하여 중앙 은행이 취하는 이자율 조절 정책의 효과를 분석할 수 있습니다.
경기 전망 보고서:

실업률과 소비자 물가 지수를 사용하여 경기 전망을 작성할 수 있습니다.
실업률의 감소는 경기 호황을 나타내며, CPI의 상승은 소비 활동의 증가를 나타낼 수 있습니다.
이러한 데이터를 기반으로 한 경기 전망은 정책 결정자, 투자자 및 기업에게 유용한 정보를 제공할 수 있습니다.
소비자 행동 및 시장 동향 보고서:

CPI 데이터를 사용하여 특정 상품 및 서비스의 가격 변화를 추적하고, 이를 통해 소비자 행동 및 시장 동향을 분석할 수 있습니다.
이자율의 변화는 대출 및 저축에 영향을 미치며, 소비자들의 구매력 및 소비 성향에 변화를 일으킬 수 있습니다.
